```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_border(B, m , prob):
    e_1 = np.sqrt(B/m)
    e_2 = np.sqrt(2*np.log(2/prob)/m)
    return max(e_1,e_2)
def check(k,e,prob):
    m = max(k/e**2, (2/e**2) * np.log(2/prob))
    return m
```

```python

m = 100000
m2 = 90000
prob= 0.1
prob3= 0.05
```

```python
y = []
y2 = [] 
y3 = []
x = range(12)
for B in x:
    y.append(get_border(B, m , prob))
    y2.append(get_border(B, m2 , prob))
    y3.append(get_border(B, m , prob3))
```

```python
plt.xlabel('$|\mathcal{B}|$')
plt.ylabel('$\epsilon_{test}$')
plt.fill_between(x, [0.012 for _ in y], y, interpolate=True, color='blue', alpha=0.1)
plt.fill_between(x, [0.012 for _ in y2], y2, interpolate=True, color='orange', alpha=0.1)
plt.fill_between(x, [0.012 for _ in y3], y3, interpolate=True, color='green', alpha=0.1)
#ax.fill_between(x, ys, where=ys<=d, interpolate=True, color='red')
plt.plot(x,y, 'o-', label='$m=100,000$ $\delta=10\%$')
plt.plot(x,y2, 'o-', label='$m=90,000$ $\delta=10\%$')
plt.plot(x,y3, 'o-', label='$m=100,000$ $\delta=5\%$')
plt.plot(x[7],y3[7], 'o', color='red')
plt.plot(x[6],y[6], 'o', color='red')
plt.plot(x[6],y2[6], 'o', color='red')
plt.text(2, 0.01, 'testing region',  fontsize=14,
        verticalalignment='top')


plt.legend()
plt.savefig('error_test.pdf')
```
